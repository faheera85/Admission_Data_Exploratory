```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import sklearn
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Admission_Predict_Ver1.1.csv")
```

```python
#Normalizing data
df['GRE Score'] = df['GRE Score']/380
df['University Rating'] = df['University Rating']/5
df['SOP']=df['SOP']/5
df['LOR ']=df['LOR ']/5
df['CGPA']=df['CGPA']/10
df['TOEFL Score']=df['TOEFL Score']/120
```

```python
#Data Visualization
df.hist(bins=60, figsize=(30,20))
plt.show()
```

```python
#Checking Normal Distribution of data for Linear Regression
nd = stats.probplot(df['TOEFL Score'], plot=plt)
plt.show()
```

```python
nd = stats.probplot(df['CGPA'], plot=plt)
plt.title('CGPA Probability')
```

```python
nd = stats.probplot(df['GRE Score'], plot=plt)
plt.title('Probability of GRE Score', color='Green', size=15)
```

```python
#Predicting Likelihood of Admission
y_df = df['Chance of Admit ']
#df.drop('Serial No.', inplace = True)
#df.drop('Chance of Admit ', inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y_df, test_size=0.2)
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)

```

```python
r = lm.score((X_test, y_test)* 100)
```
